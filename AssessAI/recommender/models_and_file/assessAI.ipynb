```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:/Datasets/exams.csv")
df
```

```python
df.isna().sum()
```

```python
df["gender"] = df.gender.apply(lambda x: 0 if x=="female" else 1)
df
```

```python
dummies = pd.get_dummies(df["race/ethnicity"], dtype="int")
```

```python
df = pd.concat([df, dummies], axis=1)
df
```

```python
second_dummies = pd.get_dummies(df["parental level of education"], dtype="int")
df = pd.concat([df, second_dummies], axis=1)
df
```

```python
df.lunch.unique()
```

```python
df["lunch"] = df["lunch"].apply(lambda x: 0 if x == 'standard' else 1)
df
```

```python
df["test preparation course"].unique()
```

```python
df["test preparation course"] = df["test preparation course"].apply(lambda x: 0 if x == 'completed' else 1)
df
```

```python
df = df.drop(["race/ethnicity", "parental level of education"], axis=1)
df
```

```python
from sklearn.cluster import KMeans
```

```python
arr = []
for i in range(2, 9):
    model = KMeans(n_clusters = i, random_state=1234)
    model.fit(df)
    arr.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(arr)
```

```python
model = KMeans(n_clusters=3, random_state=1234)
model.fit(df)
label = model.labels_
```

```python
df_labels = pd.DataFrame({"labels":label})
df = pd.concat([df, df_labels], axis=1)
df
```

```python
labels = df.labels
df = df.drop(["labels"], axis=1)
```

```python
labels
```

```python
df
```

```python
from sklearn.svm import SVC
```

```python
model_svc = SVC()
model_svc.fit(df, labels)
```

```python
model_svc.score(df, labels)
```
